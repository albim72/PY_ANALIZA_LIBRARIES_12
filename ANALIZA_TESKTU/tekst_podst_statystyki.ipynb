```python
#statystyki
word_count = len(text.split())
word_count
```

```python
char_count = len(text)
char_count
```

```python
sentence_count = text.count('.') + text.count('!') + text.count('?')
sentence_count
```

```python
bs = len(text) - text.count(' ')
bs
```
